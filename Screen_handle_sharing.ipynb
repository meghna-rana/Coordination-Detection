```python
import pandas as pd
import numpy as np
from itertools import combinations
import glob
import networkx as nx
import gc
```

```python
def removeDuplicateRecordsBasedOnNameAndId(df, name_field, id_field, STR_SEP = "/"):
    unique_daily_users = pd.Series(df.apply(
        lambda x: x[name_field] + STR_SEP + x[id_field], axis=1
    ).unique()).str.split(pat=STR_SEP, expand=True)
    unique_daily_users.columns = ["screen_name","user_id"]
    return unique_daily_users
```

# All days at once

## Concatenate the reduced files from all days

```python
files_to_process = sorted(
    glob.glob("/Users/meghnarana/Desktop/screenhandle/*.pkl.gz")
)
files_to_process
#elections20*.pkl.gz
```

```python
all_users = []
for daily_file in files_to_process:
    print(daily_file)
    daily_tw = pd.read_pickle(daily_file)
    unique_daily_users = removeDuplicateRecordsBasedOnNameAndId(
        daily_tw, name_field="user.screen_name", id_field="user.id_str"
    )
    all_users.append(unique_daily_users)
```

```python
del all_users
```

```python
all_users = pd.concat(all_users)
```

## Prepare a dataframe of screen_name shared by multiple users

```python
circular_user = all_users.groupby("screen_name").agg(
    {"user_id":["nunique","unique"]}
)
circular_user.columns = ["nunique_usr","users"]
circular_user.info()
circular_user.head()
circular_user.sort_values("nunique_usr",ascending=False).head()
```

## Create projection over users

```python
user_proj = pd.DataFrame(
    np.concatenate(
        circular_user[circular_user.nunique_usr>1].users.apply(combinations,r=2).apply(list)
    ),
    columns=["source","target"]
).groupby(["source","target"]).size().rename("weight")
```
